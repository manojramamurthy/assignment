```python
def predict(image_path):
    img = cv2.imread(image_path)
    height = img.shape[0]
    width = img.shape[1]
    model = YOLO('weights2/best.pt')
    results = model([image_path])
    class_name = 'Coin'
    
    for result in results:
        boxes = result.boxes.cpu().numpy()
        
        xyxys = boxes.xyxy
        print(xyxys)
        for xyxy in xyxys:
            cv2.rectangle(
                img,
                (int(xyxy[0]), int(xyxy[1])),
                (int(xyxy[2]), int(xyxy[3])),
                (0, 255, 0),
                thickness=2
            )
        
            font_scale = min(1, max(3, int(width/500)))
            font_thickness = min(2, max(10, int(width/50)))
            
            p1, p2 = (int(xyxy[0]), int(xyxy[1])), (int(xyxy[2]), int(xyxy[3]))
            # Text width and height
            tw, th = cv2.getTextSize(
                class_name,
                0, fontScale=font_scale, thickness=font_thickness
            )[0]
            p2 = p1[0] + tw, p1[1] + -th - 10
            
            cv2.rectangle(
                img,
                p1, p2,
                color=(0, 255, 0),
                thickness=-1
            )
            
            cv2.putText(
                img,
                class_name,
                (int(xyxy[0]) + 1 , int(xyxy[1]) - 10),
                cv2.FONT_HERSHEY_SIMPLEX,
                font_scale,
                (255, 255, 255),
                font_thickness
            )
    
    return img
```

```python
# Directory containing the input images
input_dir = "data/coin-dataset/test/images"
# input_dir = "data/coin-dataset/val/images"
# input_dir = "data/coin-dataset/test/images"

# Loop over the input images
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        image_path = os.path.join(input_dir, filename)
        image = predict(image_path)
        plt.imsave('predict/'+filename, image)
        # plt.imshow(image)
```
